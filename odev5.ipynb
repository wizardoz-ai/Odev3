```python
# Önce Kütüphaneleri İçeri Aktaralım

import yfinance as yf
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# İstediğimiz bir hisse senedi verisini almak için:
symbol = 'NVDA'  # Burada istediğimiz hisse sembolünü gireceğiz
data = yf.download(symbol, start="2022-11-01", end="2023-11-01")
```

```python
# 20 günlük SMA ve 50 günlük EMA hesaplayalım
data['SMA_20'] = data['Close'].rolling(window=20).mean()
data['EMA_50'] = data['Close'].ewm(span=50, adjust=False).mean()
```

```python
# Fiyat ve Hareketli Ortalamaları Görselleştirelim

plt.figure(figsize=(14, 7))
plt.plot(data['Close'], label='Close Price')
plt.plot(data['SMA_20'], label='20-Day SMA')
plt.plot(data['EMA_50'], label='50-Day EMA')
plt.title(f'{symbol} Fiyat Hareketleri ve Hareketli Ortalamalar')
plt.xlabel('Tarih')
plt.ylabel('Fiyat (USD)')
plt.legend()
plt.show()
```
